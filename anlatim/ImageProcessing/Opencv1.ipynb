## Resim üzerinde operasyon

```python
img4 = img.copy()
cv2_imshow(img4[0:100,0:100])
```

```python
img4[0:100,0:100] = [255,255,255]
cv2_imshow(img4)
```

```python
img5 = img.copy()
logo = cv2.imread("mainlogo.png")

satir,sutun,kanal = logo.shape
roi = img5[0:satir,0:sutun]

logogri = cv2.cvtColor(logo,cv2.COLOR_BGR2GRAY)

ret,mask = cv2.threshold(logogri,220,255,cv2.THRESH_BINARY_INV)

mask_inv = cv2.bitwise_not(mask)

img5_bg = cv2.bitwise_and(roi,roi,mask=mask_inv)
img5_fg = cv2.bitwise_and(logo,logo,mask = mask)

dst = cv2.add(img5_bg,img5_fg)
img5[0:satir,0:sutun] = dst
cv2_imshow(img5)
```

## Thresold

```python
book = cv2.imread("kitap.jpg")
cv2_imshow(book)
```

```python
retval, threshold = cv2.threshold(book,10,255,cv2.THRESH_BINARY)
threshold
```

```python
gri = cv2.cvtColor(book,cv2.COLOR_BGR2GRAY)
retval, threshold = cv2.threshold(gri,10,255,cv2.THRESH_BINARY)
threshold
```

```python
gri = cv2.cvtColor(book,cv2.COLOR_BGR2GRAY)
threshold = cv2.adaptiveThreshold(gri,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,115,1)
threshold
```

## Filtering

```python
renkResim = cv2.imread("renkler.jpg")
cv2_imshow(renkResim)
```

```python
import numpy as np
hsv = cv2.cvtColor(renkResim,cv2.COLOR_BGR2HSV)
lower_red = np.array([30,150,50])
upper_red = np.array([255,255,180])

maske = cv2.inRange(hsv,lower_red,upper_red)
sonuc = cv2.bitwise_and(renkResim,renkResim,mask=maske)
cv2_imshow(sonuc)
```

## Blur and Smooting

```python
kedi = cv2.imread("resim.jpg")
cv2_imshow(kedi)
```

```python
cekirdek = np.ones((15,15))/225
cv2_imshow(cv2.filter2D(kedi,0,cekirdek))
```

```python
kedi[150:250,450:500] =  cv2.GaussianBlur(kedi[150:250,450:500],(15,15),0)
cv2_imshow(kedi)
```

```python
 cv2.GaussianBlur(kedi[150:300,450:600],(15,15),0)
```

```python
bilateral = cv2.bilateralFilter(kedi,25,75,75)
cv2_imshow(bilateral)
```
