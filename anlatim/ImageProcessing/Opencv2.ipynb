```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
yuzcas = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
gozcas = cv2.CascadeClassifier("haarcascade_eye.xml")
```

```python
img = cv2.imread("yuzler1.jpg")
```

```python
gri = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
yuzler = yuzcas.detectMultiScale(gri,1.3,8)
```

```python
img1 = img.copy()
for (x,y,w,h) in yuzler:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
cv2_imshow(img1)
```

```python
img1 = img.copy()
for (x,y,w,h) in yuzler:
  cv2.rectangle(img1,(x,y),(x+w,y+h),(0,0,255),2)
  roi_gri = gri[y:y+h,x:x+w]
  roi = img1[y:y+h,x:x+w]
  gozler = gozcas.detectMultiScale(roi_gri,1.2,3)
  for (ex,ey,ew,eh) in gozler:
    cv2.rectangle(roi,(ex,ey),(ex+ew,ey+eh),(0,255,0),-1)
cv2_imshow(img1)
```

```python
yuzcas = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
gozcas = cv2.CascadeClassifier("haarcascade_eye.xml")
def yuzTespit(adres):
  img = cv2.imread(adres)
  gri = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  yuzler = yuzcas.detectMultiScale(gri,1.3,5)
  img1 = img.copy()
  for (x,y,w,h) in yuzler:
    cv2.rectangle(img1,(x,y),(x+w,y+h),(0,0,255),2)
    roi_gri = gri[y:y+h,x:x+w]
    roi = img1[y:y+h,x:x+w]
    gozler = gozcas.detectMultiScale(roi_gri,1.2,3)
    for (ex,ey,ew,eh) in gozler:
      cv2.rectangle(roi,(ex,ey),(ex+ew,ey+eh),(0,255,0),-1)
  cv2_imshow(img1)
```

```python
liste = ["a-milli-kadin-voleybol-takimi-1591893.jpg",
         "takimimgs.jpg",
         "real.jpg",
         "yuzler1.jpg"]

for item in liste:
  yuzTespit(item)
```
