```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fig, ax = plt.subplots()
plt.show()
```

```python
adres1 = "/content/drive/MyDrive/KDKEgitim/Egzersizler/data/seattle_weather.csv"
adres2 = "/content/drive/MyDrive/KDKEgitim/Egzersizler/data/austin_weather.csv"
df1 = pd.read_csv(adres1)
df2 = pd.read_csv(adres2)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.columns
```

```python
ista1 = df1.groupby("DATE")["MLY-TAVG-NORMAL"].mean()
ista2 = df2.groupby("DATE")["MLY-TAVG-NORMAL"].mean()
ista1.index=["Ocak","Şubat","Mart","Nisan","Mayıs","Haziran","Temmuz","Ağustos","Eylül","Ekim","Kasım","Aralık"]
ista2.index=["Ocak","Şubat","Mart","Nisan","Mayıs","Haziran","Temmuz","Ağustos","Eylül","Ekim","Kasım","Aralık"]
```

## çoklu grafik gösterimi

```python
fig, ax = plt.subplots()
ax.plot(ista1.index,ista1)
ax.plot(ista2.index,ista2)
plt.show()
```

### marker

```python
fig, ax = plt.subplots()
ax.plot(ista1.index,ista1,marker="v")
ax.plot(ista2.index,ista2,marker="*")
plt.show()
```

### linestyle

```python
fig, ax = plt.subplots()
ax.plot(ista1.index,ista1,marker="v",linestyle="--")
ax.plot(ista2.index,ista2,marker="*",linestyle="None")
plt.show()
```

### color

```python
fig, ax = plt.subplots()
ax.plot(ista1.index,ista1,marker="v",linestyle="--",color="r")
ax.plot(ista2.index,ista2,marker="*",linestyle="None",color="b")
plt.show()
```

#### ax etiketleri

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(ista1.index,ista1,marker="v",linestyle="--",color="r")
ax.plot(ista2.index,ista2,marker="*",linestyle="None",color="b")
ax.set_xlabel("Ay Bilgileri")
ax.set_ylabel("Ortalama Sıcaklıklar")
plt.show()
```

```python
ista1 = df1.groupby("DATE")[[f"MLY-GRDD-BASE{x}" for x in range(40,80,10)]].mean()
```

```python
ista1.index=["Ocak","Şubat","Mart","Nisan","Mayıs","Haziran","Temmuz","Ağustos","Eylül","Ekim","Kasım","Aralık"]
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(ista1.index,ista1["MLY-GRDD-BASE40"],marker="v",linestyle="--",color="r")
ax.plot(ista1.index,ista1["MLY-GRDD-BASE50"],marker="o",linestyle="--",color="b")
ax.plot(ista1.index,ista1["MLY-GRDD-BASE60"],marker="*",linestyle="--",color="y")
ax.plot(ista1.index,ista1["MLY-GRDD-BASE70"],marker="x",linestyle="--",color="g")
plt.show()
```

```python
fig, ax = plt.subplots(2,2,figsize=(10,10))
ax[0,0].plot(ista1.index,ista1["MLY-GRDD-BASE40"],marker="v",linestyle="--",color="r")
ax[0,1].plot(ista1.index,ista1["MLY-GRDD-BASE50"],marker="o",linestyle="--",color="b")
ax[1,0].plot(ista1.index,ista1["MLY-GRDD-BASE60"],marker="*",linestyle="--",color="y")
ax[1,1].plot(ista1.index,ista1["MLY-GRDD-BASE70"],marker="x",linestyle="--",color="g")
plt.show()
```
