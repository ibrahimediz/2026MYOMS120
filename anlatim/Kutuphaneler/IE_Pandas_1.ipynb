### scikit learn kütüphanesinde bulunan hazır veri setleri nasıl alınır

```python
from sklearn.datasets import load_wine
veri = load_wine()
df = pd.DataFrame(data=veri.data)
df.columns = veri.feature_names
df["target"] = veri.target

```

```python
df.head()
```

```python
df.groupby("target")["alcohol"].mean()
```

```python
df.groupby("target")["alcohol"].max()
```

```python
df.groupby("target")["alcohol"].agg([max,min,sum,np.mean])
```

```python
df.groupby(["target","color_intensity"])[["ash","magnesium"]].agg([max,min,sum,np.mean])
```

```python
df.pivot_table(values=["ash","magnesium"],index=["target","color_intensity"],aggfunc=[max,min,sum,np.mean])
```

```python
istatistik = df.pivot_table(values="alcohol",index="target",columns="ash",fill_value=0)
```
