1. Yukarıkdaki kod bloğunu kullanarak dataframe oluşturun
2. ***target*** sütununa ait verilerin **dağılım** grafikle gösterimini sağlayınız (hist)
3. ***target*** sütununa göre gruplanmış verilerin ***alcohol*** sütununa göre dağılımını gösteren grafiği çizdiriniz (hist,bar)
4. ***target*** sütununa göre ***magnesium*** değerini gösteren **noktasal** grafiği çizdiriniz (scatter)

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
veri = load_wine()
df = pd.DataFrame(data=veri.data)
df.columns = veri.feature_names
df["target"] = veri.target
````

````python
seri1 =  df.groupby("target")["alcohol"].count()
seri1.index = ["red","white","rose"]
seri1.plot(kind="bar")
plt.show()

````

````python
df.groupby("target")["alcohol"].hist()
plt.legend(["red","white","rose"]);
````

````python
df[df["target"]==0]["alcohol"].hist()
df[df["target"]==1]["alcohol"].hist()
df[df["target"]==2]["alcohol"].hist()
plt.legend(["red","white","rose"]);
````

````python
df.plot(y="target",x="magnesium",kind="scatter");
````
