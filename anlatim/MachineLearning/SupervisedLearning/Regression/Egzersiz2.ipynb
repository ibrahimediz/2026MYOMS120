```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df =  pd.read_csv("housing.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df["total_bedrooms"].fillna(df["total_bedrooms"].mean(),inplace=True)
```

```python
np.random.seed(2)
rowNum = np.random.randint(1,df.shape[0])
realTestVal = df.iloc[rowNum,:]
df.drop(rowNum,axis=0,inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
realTestVal
```

```python
X=df.drop("median_house_value",axis=1)
y=df["median_house_value"]
```

```python
df.ocean_proximity.value_counts()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[("encoder",OneHotEncoder(),[-1])],remainder="passthrough")
X = ct.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_scaled = sc.transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
# pd.DataFrame(X_train_scaled)
```

```python
pd.DataFrame(X_train_scaled).isna().sum()
```

```python
from sklearn.svm import SVR
model = SVR(kernel="linear")
model.fit(X_train_scaled,y_train)
```

```python
y_preds = model.predict(X_test_scaled)
```

```python
from math import sqrt
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
r2_score(y_test,y_preds),sqrt(mean_squared_error(y_test,y_preds)),mean_absolute_error(y_test,y_preds)
```
