```python
def metin_temiz(metin):

  # noktalama işaretlerini temizle
  metin = re.sub(r'[^\w\s]','',metin)
  # fazla boşlukları kaldırma
  metin = re.sub(' +',' ',metin)
  print("metin_temiz => :",metin)
  # küçük harfe dönüştürme
  metin = metin.lower()
  return metin
metin_temiz(cumle)
```

```python
def cumle_cikar(metin):
  return metin.split(".")
```

```python
def kelime_parcala(metin):
  # cümlelere ayırma
  cumleler = sent_tokenize(metin)
  # kelimelere ayırma
  kelimeler = [word_tokenize(cumle) for cumle in cumleler]
  return kelimeler
kelime_parcala(metin_temiz(cumle))
```

```python
def durdurma_kelimeleri_kaldır(kelimeler):
  # Türkçe durdurma kelimelerini al
  dur_kelimeler = set(stopwords.words("turkish"))
  kelimeler = [kelime for kelime in kelimeler[0] if kelime not in dur_kelimeler]
  return kelimeler
durdurma_kelimeleri_kaldır(kelime_parcala(metin_temiz(cumle)))
```

```python
def kok_bul(kelimeler):
  # Türkçe Kök Bulucu
  kok_bulucu = TurkishStemmer()
  # kelimeleri köklerine indirgeme
  kokler = [kok_bulucu.stemWord(kelime) for kelime in kelimeler]
  return kokler
kok_bul(durdurma_kelimeleri_kaldır(kelime_parcala(metin_temiz(cumle))))
```

```python
def cumleleri_hazirla(cumleler):
    hazir_cumlelerList = []
    # print("cumleleri_hazirla 1=> :",cumleler)
    hazir_cumleler = metin_temiz(cumleler)
    # print("cumleleri_hazirla 2=> :",hazir_cumleler)
    hazir_cumleler = kelime_parcala(hazir_cumleler)
    # print("cumleleri_hazirla 3=> :",hazir_cumleler)
    hazir_cumleler = durdurma_kelimeleri_kaldır(hazir_cumleler)
    # print("cumleleri_hazirla 4=> :",hazir_cumleler)
    hazir_cumleler = kok_bul(hazir_cumleler)
    # print("cumleleri_hazirla 5=> :",hazir_cumleler)
    hazir_cumlelerList.append(hazir_cumleler)
    return hazir_cumlelerList
cumleleri_hazirla(cumle)
```

##Veri Seti Sınıfı Oluşturma

```python
from torch.utils.data import Dataset,DataLoader

class MetinDataset(Dataset):
    def __init__(self,encoded_metin):
      self.data = encoded_metin

    def __len__(self):
      return len(self.data)

    def __getitem__(self,index):
      return self.data[index]

```

```python

def metin_encode(cumleler):
    vectorizer = CountVectorizer() # BoW
    # vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(cumleler)
    encoded_metin = X.toarray()
    return encoded_metin,vectorizer

```
