```python
import torch.nn as nn
import torch.optim as optim
class KararAnalyticsCNN(nn.Module):
    def __init__(self, metin_size, embed_dim):
      super().__init__()
      self.embedding = nn.Embedding(metin_size,embed_dim)
      self.conv = nn.Conv1d(embed_dim,embed_dim,kernel_size=3,stride=1,padding=1)
      self.fc = nn.Linear(embed_dim,2)

    def forward(self,text):
      embedded = self.embedding(text).permute(0,2,1)
      conved = nn.functional.relu(self.conv(embedded))
      conved = conved.mean(dim=2)
      return self.fc(conved)
```

```python
metin = ["Bu","akşam","ölürüm","beni","kimse","tutamaz"]
kel_to_idx = {kelime:i for i,kelime in enumerate(metin) }
kel_to_idx
```

```python
embed_dim = 10
metin_size = len(kel_to_idx)
```

```python
ornekler = [
    ("Seni çöpe atacağım poşete yazık.".split(),0),
 ("Bilsem ki bir daha hiç dönmeyecek".split(),1)]

model = KararAnalyticsCNN(metin_size,embed_dim)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(),lr=0.1)
```

```python
import torch
for epoch in range(1000):
  for cumle, label in ornekler:
    model.zero_grad()
    cumle = torch.LongTensor([kel_to_idx.get(w,0) for w in cumle]).unsqueeze(0)
    outputs = model(cumle)
    label = torch.LongTensor([int(label)])
    loss = criterion(outputs,label)
    loss.backward()
    optimizer.step()
```

```python
for ornek in ornekler:
  input_tensor = torch.tensor([kel_to_idx.get(w,0) for w in cumle],dtype=torch.long).unsqueeze(0)
  outputs = model(input_tensor)
  _, predicted = torch.max(outputs.data,1)
  tespit = "Pozitif" if predicted.item() == 1 else "Negatif"
  print(ornek,tespit)
```
