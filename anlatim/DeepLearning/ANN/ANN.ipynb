## BackPropogation

```python
giris_veri = torch.tensor(
    [[0.4680, 0.1450, 0.1029, 0.3760, 0.5813, 0.4607, 0.8840, 0.9123, 0.6441,
         0.5628],
        [0.7681, 0.2212, 0.9511, 0.4000, 0.2130, 0.7685, 0.8511, 0.2762, 0.5940,
         0.8605],
    [0.4178, 0.9666, 0.2885, 0.1685, 0.2009, 0.7904, 0.1492, 0.2379, 0.9642,
        0.9226]]
)
etiket = torch.tensor([[1],[0],[1]])
```

```python
model = nn.Sequential(
    nn.Linear(10,8),
    nn.Linear(8,4),
    nn.Linear(4,1),
     nn.Sigmoid()
)
cikti = model(giris_veri)
print(cikti)
```

```python
criterion = nn.BCELoss()
loss = criterion(cikti.double(),etiket.double())
loss.backward()   #Back Propogation
```
