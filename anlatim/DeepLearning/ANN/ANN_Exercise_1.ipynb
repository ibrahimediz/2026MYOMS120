## Modelin Oluşturulması

```python
model = nn.Sequential(
    nn.Linear(4,16),
    nn.Linear(16,8),
    nn.Linear(8,1),
    nn.Sigmoid()
)
```

```python
import torch.optim as optim
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(),lr=0.01,momentum=0.94)
```

```python
num_epochs = 10
for epoch in range(num_epochs):
  for data in dataloader:
    optimizer.zero_grad()
    feature,target = data
    pred = model(feature)
    # print(pred)
    # print(target)
    loss = criterion(pred,target.unsqueeze(1).float())
    loss.backward()
    optimizer.step()
```

```python
import torchmetric
metric = torchmetrics.Accuracy(task="binary")
num_epochs = 10
for epoch in range(num_epochs):
  for data in dataloader:
    optimizer.zero_grad()
    feature,target = data
    pred = model(feature)
    # print(pred)
    # print(target)
    loss = criterion(pred,target.unsqueeze(1).float())
    acc = metric(pred,target.unsqueeze(1).float())
    loss.backward()
    optimizer.step()
acc = metric.compute()
```

```python
from sklearn.model_selection import train_test_split
df = pd.read_csv("water_potability.csv")
df_train,df_test = train_test_split(df,test_size=0.2,random_state=42)
# df_train.to_csv("water_train.csv",index=False)
# df_test.to_csv("water_test.csv",index=False)
df_train.shape
df_test.shape
```

```python
from torch.utils.data import Dataset
class WaterDataset(Dataset):
  def __init__(self,csv_path):
    super().__init__()
    df = pd.read_csv(csv_path)
    self.data = df.to_numpy()

  def __len__(self):
    return self.data.shape[0]

  def __getitem__(self,idx):
    feature = self.data[idx,:-1]
    label =self.data[idx,-1]
    return feature,label

dataset_train = WaterDataset("water_train.csv")
dataset_test = WaterDataset("water_test.csv")
```

```python
next(iter(dataset_train))
```

```python
from torch.utils.data import DataLoader
dataloader_train = DataLoader(
    dataset_train,
    batch_size=5,
    shuffle=True
)
```

## sequential model tanimi

```python
model = nn.Sequential(
    nn.Linear(9,16),
    nn.ReLU(),
    nn.Linear(16,8),
    nn.ReLU(),
    nn.Linear(8,1),
    nn.Sigmoid()
)
```

```python
class Net(nn.Module):
  def __init__(self):
    super(Net,self).__init__()
    self.fc1 =nn.Linear(9,32)
    self.fc2 =nn.Linear(32,16)
    self.fc3 =nn.Linear(16,8)
    self.fc4 =nn.Linear(8,4)
    self.fc5 =nn.Linear(4,1)
    self.double()

  def forward(self,x):
    x = nn.functional.relu(self.fc1(x))
    x = nn.functional.relu(self.fc2(x))
    x = nn.functional.relu(self.fc3(x))
    x = nn.functional.relu(self.fc4(x))
    x = nn.functional.sigmoid(self.fc5(x))
    return x

```

## Eğitim Döngüsü

```python
# %pip install torchmetrics
```

```python
from torchmetrics import Accuracy

acc = Accuracy(task="binary")
acc.reset()
net = Net()
criterion = nn.BCELoss()
optimizer = optim.Adagrad(net.parameters(),lr=0.01)
for epoch in range(10):
  for features, labels in dataloader_train:
    optimizer.zero_grad()
    outputs = net(features)
    preds = (outputs>=0.5).float()
    acc(preds,labels.view(-1,1))
    loss = criterion(outputs,labels.view(-1,1))
    loss.backward()
    optimizer.step()

accuracy = acc.compute()
accuracy
```
