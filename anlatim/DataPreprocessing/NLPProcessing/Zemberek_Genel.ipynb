```python
cumle = "Yarın kar yağacak."
from zemberek import TurkishSpellChecker
sc =TurkishSpellChecker(morph)
analiz = morph.analyze_sentence(cumle)
sonuc = morph.disambiguate(cumle,analiz)
for item in analiz:
  print(f"Kelime = {item.inp}")
  for s in item:
    print(s.format_string())
print("#"*30)
for s in sonuc.best_analysis():
  print(s.format_string())
  # print(s.get_stem())
  print(sc.suggest_for_word(s.get_stem()))
```

```python
from zemberek import TurkishTokenizer
tokenizer = TurkishTokenizer.DEFAULT

tokens = tokenizer.tokenize("Çöpler var anladın mı")
for token in tokens:
  print("İçerik=",token.content)
  print("Type=",token.type_.name)
  print("Start=",token.start)
  print("Stop=",token.end)
```
