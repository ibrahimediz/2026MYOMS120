# Regex
## Metakarakterler:

* .: Herhangi bir karakteri temsil eder (newline hariç).
* ^: Satır başını belirtir.
* $: Satır sonunu belirtir.
* *: Bir ifadenin sıfır veya daha fazla tekrarını belirtir.
* +: Bir ifadenin bir veya daha fazla tekrarını belirtir.
* ?: Bir ifadenin sıfır veya bir kez tekrarını belirtir.
* \: Bir sonraki karakterin özel bir anlam taşımamasını sağlar.
  
## Karakter Sınıfları:

* [abc]: a, b veya c karakterlerinden herhangi birini eşleştirir.
* [^abc]: a, b veya c dışındaki herhangi bir karakteri eşleştirir.
* [a-z]: a'dan z'ye kadar olan tüm küçük harfleri eşleştirir.
* [A-Z]: A'dan Z'ye kadar olan tüm büyük harfleri eşleştirir.
* [0-9]: 0'dan 9'a kadar olan tüm rakamları eşleştirir.
  
## Özel Düzenleyiciler:

* \d: Sayısal bir karakteri eşleştirir.
* \D: Sayısal olmayan bir karakteri eşleştirir.
* \s: Bir boşluk karakterini eşleştirir.
* \S: Boşluk olmayan bir karakteri eşleştirir.
* \w: Bir kelime karakterini eşleştirir (harf, rakam, alt çizgi).
* \W: Kelime olmayan bir karakteri eşleştirir.
  
## Gruplar ve Yakalama:

* () : Grup oluşturur ve eşleşen kısmı yakalar.
* (?:) : Yakalamayan grup oluşturur.
* \1, \2, ...: İlk, ikinci, ... grupları referans alır.

## Özel Karakterler:

* \t: Sekme karakterini temsil eder.
* \n: Yeni satır karakterini temsil eder.

# Türkçe karakterlere uygun küçük harfe dönüştürme

````python
import re

def tr_upper(self):
    self = re.sub(r"i", "İ", self)
    self = re.sub(r"ı", "I", self)
    self = re.sub(r"ç", "Ç", self)
    self = re.sub(r"ş", "Ş", self)
    self = re.sub(r"ü", "Ü", self)
    self = re.sub(r"ğ", "Ğ", self)
    self = self.upper() # for the rest use default upper
    return self


def tr_lower(self):
    self = re.sub(r"İ", "i", self)
    self = re.sub(r"I", "ı", self)
    self = re.sub(r"Ç", "ç", self)
    self = re.sub(r"Ş", "ş", self)
    self = re.sub(r"Ü", "ü", self)
    self = re.sub(r"Ğ", "ğ", self)
    self = self.lower() # for the rest use default lower
    return self
````

````python
from string import punctuation
liste = []
for item in metin:
  if not item.isdigit():
    if item not in punctuation:
      liste.append(tr_lower(item))
metinAlt1 = "".join(liste)
metinAlt1
````

````python
from nltk import word_tokenize
liste = word_tokenize(metinAlt1)
liste[:5]
````

````python
corpus = [sozluk.doc2bow([item]) for item in liste]
corpus
````
