## 1. Karşılaştırma Operatörleri

Karar verirken durumları karşılaştırırız. Sonuç her zaman `True` (Doğru) veya `False` (Yanlış) olur.

*   `==` : Eşit mi?
*   `!=` : Eşit değil mi?
*   `>` : Büyük mü?
*   `<` : Küçük mü?
*   `>=` : Büyük veya eşit mi?
*   `<=` : Küçük veya eşit mi?

```python
print(5 == 5)   # True
print(5 == 10)  # False
print(10 > 2)   # True
```

## 2. Mantıksal Operatörler

Birden fazla koşulu birleştirmek için kullanılır.

*   `and` (VE): İkisi de doğruysa True.
*   `or` (VEYA): En az biri doğruysa True.
*   `not` (DEĞİL): Tersi.

```python
yas = 20
ehliyet = True

# Araba kullanabilir mi?
# Hem yaşı 18'den büyük OLMALI HEM DE ehliyeti OLMALI
if yas >= 18 and ehliyet == True:
    print("Araba kullanabilirsin.")
else:
    print("Araba kullanamazsın.")
```

## 3. If - Elif - Else Yapısı

Programın akışını yönlendiren ana bloklardır.

```python
puan = 75

if puan >= 90:
    print("Harf Notu: AA")
elif puan >= 80:
    print("Harf Notu: BA")
elif puan >= 70:
    print("Harf Notu: BB")
else:
    print("Daha çok çalışmalısın.")
```

## 4. Döngüler (Loops)

Bilgisayarların en sevdiği şey: Tekrar etmek. Aynı işi binlerce kez hatasız yapabilirler.

### 4.1 While Döngüsü
Bir koşul doğru olduğu sürece dönmeye devam eder.

```python
sayac = 0
while sayac < 5:
    print("Sayaç şu an: ", sayac)
    sayac = sayac + 1
print("Döngü bitti.")
```

### 4.2 For Döngüsü
Belirli bir listenin veya aralığın üzerinde gezinmek için kullanılır. En sık kullanılan döngüdür.

```python
# range(baslangic, bitis, artis_miktari)
print("1'den 10'a kadar tek sayılar:")
for i in range(1, 10, 2):
    print(i)
```

## 5. Break ve Continue

*   `break`: Döngüyü anında kırar ve çıkar.
*   `continue`: O anki turu atlar ve başa döner.

```python
print("Break örneği:")
for i in range(10):
    if i == 5:
        break # 5 olunca çık
    print(i)

print("\nContinue örneği:")
for i in range(5):
    if i == 2:
        continue # 2'yi atla
    print(i)
```
