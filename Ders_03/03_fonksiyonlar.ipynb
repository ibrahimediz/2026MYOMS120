## 1. Fonksiyon Nedir?

Belirli bir görevi yapan kod bloğudur. Matematiği hatırla: $f(x) = x^2$. Girdi alır, işlem yapar, çıktı verir.

```python
def selamla():
    print("Merhaba!")

# Fonksiyonu tanımladık ama çalışmadı. Çağırmamız lazım:
selamla()
selamla()
```

## 2. Parametreler ve Argümanlar

Fonksiyonlara dışarıdan veri gönderebiliriz.

```python
def kisi_selamla(isim):
    print(f"Merhaba sevgili {isim}, hoşgeldin.")

kisi_selamla("Ahmet")
kisi_selamla("Ayşe")
```

## 3. Return (Değer Döndürme)

Fonksiyon sadece ekrana yazı yazmaz, işlemi yapıp sonucu bize geri verebilir. Bu sonucu başka yerde kullanabiliriz.

```python
def topla(a, b):
    return a + b

sonuc = topla(5, 10)
print("Sonuç:", sonuc)
print("Sonucun karesi:", sonuc * sonuc)
```

## 4. Scope (Kapsam) Kavramı

*   **Global Değişken:** Her yerden erişilebilir, ana programda tanımlanır.
*   **Yerel (Local) Değişken:** Sadece fonksiyon içinde yaşar, fonksiyon bitince silinir.

```python
global_degisken = "Ben her yerdeyim"

def deneme():
    yerel_degisken = "Ben sadece buradayım"
    print(global_degisken)
    print(yerel_degisken)

deneme()
# print(yerel_degisken) # HATA! Çünkü dışarıdan erişilemez.
```

## 5. Uygulama: Gelişmiş Hesap Makinesi

Öğrendiklerimizi birleştirelim.

```python
def hesap_makinesi(sayi1, sayi2, islem):
    if islem == "+":
        return sayi1 + sayi2
    elif islem == "-":
        return sayi1 - sayi2
    elif islem == "*":
        return sayi1 * sayi2
    elif islem == "/":
        if sayi2 == 0:
            return "Hata: Sıfıra bölünemez!"
        return sayi1 / sayi2
    else:
        return "Geçersiz İşlem"

print(hesap_makinesi(10, 2, "*"))
print(hesap_makinesi(10, 0, "/"))
```
