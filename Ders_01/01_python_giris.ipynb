## 4. Veri Tipleri (Data Types)

Python'da veriler farklı türlerde olabilir. Temel ve Koleksiyon tipi olarak ayırabiliriz:

### 4.1. Temel Tipler (Primitive Types)
1.  **Integer (int):** Tam sayılar (5, -10, 0)
2.  **Float:** Ondalıklı sayılar (3.14, -0.001, 2.0)
3.  **String (str):** Metin ifadeleri ("Merhaba", 'Python')
4.  **Boolean (bool):** Mantıksal değerler (True, False)

```python
sayi = 10
pi = 3.14
metin = "Python Dersleri"
dogru_mu = True

print(type(sayi))
print(type(pi))
print(type(metin))
print(type(dogru_mu))
```

### 4.2. Koleksiyon Tipleri (Kısaca)
Bu konuyu 4. derste detaylı işleyeceğiz ama şimdiden bilmekte fayda var:
*   **List (Liste):** `[1, 2, 3]` -> Değiştirilebilir, sıralı liste.
*   **Tuple (Demet):** `(1, 2, 3)` -> Değiştirilemez (sabit) liste.
*   **Dictionary (Sözlük):** `{'ad': 'Ali', 'yas': 20}` -> Anahtar-Değer ikilisi.
*   **Set (Küme):** `{1, 2, 3}` -> Eşsiz elemanlar, sırasız.

```python
# Basit örnekler
liste = [10, 20, 30]
sozluk = {"elma": "kırmızı", "muz": "sarı"}

print(liste)
print(sozluk["elma"])
```

## 5. Tip Dönüşümleri (Type Casting)

Bazen metin olarak gelen bir sayıyı matematiksel işleme sokmak isteyebiliriz. Bunun için tip dönüşümü yaparız.

```python
# String'i Integer'a çevirme
yazi_sayi = "50"
# print(yazi_sayi + 10) # Bu HATA verir! Çünkü yazı ile sayı toplanmaz.

gercek_sayi = int(yazi_sayi)
print(gercek_sayi + 10) # Şimdi çalışır sonuç 60 olur.

# Sayıyı String'e çevirme
notum = 95
mesaj = "Notum: " + str(notum)
print(mesaj)
```

## 6. Kullanıcıdan Veri Alma (Input)

Programların dinamik olması için dışarıdan veri almamız gerekir.

```python
ad = input("Lütfen adınızı giriniz: ")
dogum_yili = input("Doğum yılınız: ")

# input her zaman String verir! Yaş hesabı için int'e çevirmeliyiz.
yas = 2024 - int(dogum_yili)

print("Merhaba " + ad + ", yaşınız: " + str(yas))
```
